# Correlation check

```python
numeric_credit_df = credit_card.select_dtypes(include=['int64', 'float64'])
plt.matshow(credit_card.corr(),cmap='Greens')
plt.xticks(range(len(numeric_credit_df.columns)), numeric_credit_df.columns)
plt.yticks(range(len(numeric_credit_df.columns)), numeric_credit_df.columns)
plt.colorbar()
plt.show()

print(credit_card.corr())
```

 Based on this correlation matrix, we notice that Balance appears to be strong correlated with Limit, Rating and moderately correlated with Income. 

```python
# How well Income predicts Balance

lm2 = smf.ols('Balance ~ Income', data=credit_card).fit()
lm2.summary()
```

```python
# How well Limit predicts Sales

lm3 = smf.ols('Balance ~ Limit', data=credit_card).fit()
lm3.summary()
```

```python
# How well Newspaper predicts Sales

lm4 = smf.ols('Balance ~ Rating', data=credit_card).fit()
lm4.summary()
```

```python
# How well Age, Education, Ethnicity, Student and Married predicts Sales

lm_irrelevant = smf.ols('Balance ~ Age + Education + Ethnicity + Student + Married', data=credit_card).fit()
lm_irrelevant.summary()
```

```python
# How well Income + Limit + Rating + Cards predicts Balance
lm_relevant = smf.ols('Balance ~ Income + Limit + Rating + Cards', data = credit_card).fit()
lm_relevant.summary()
```

We notice that p_value of cards is somewhat bigger than the accepted value (5% = 0.05).

# Applying Linear Regression method

```python
cc_copy = credit_card
cc_copy['Student']=pd.get_dummies(cc_copy['Student'])
cc_copy['Gender']=pd.get_dummies(cc_copy['Gender'])
cc_copy['Married']=pd.get_dummies(cc_copy['Married'])
ethnicity = pd.get_dummies(cc_copy['Ethnicity'])
```

```python
ethnicity = np.argmax(ethnicity.values,axis=1)
ethnicity
```

```python
cc_copy['Ethnicity']=ethnicity
```

```python
plt.matshow(cc_copy.corr(),cmap='Greens')
plt.xticks(range(len(cc_copy.columns)),cc_copy.columns)
plt.yticks(range(len(cc_copy.columns)),cc_copy.columns)
plt.colorbar()
plt.show()
```

We notice that Age, Education, Gender, Married, Ethnicity and Student are not relevant in this dataset. Thus, we could drop these columns in order to make more accurate prediction.

```python
cc_copy.drop(columns=['Age','Education','Gender','Married','Ethnicity','Student'],inplace=True)
```

```python
cc_copy.head()
```

```python
#X = credit_card['Rating'].values.reshape(-1,1)
y = credit_card['Balance'].values.reshape(-1,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(cc_copy, y, test_size=0.2, random_state=67)
```

```python
regressor = LinearRegression(normalize=True)
regressor.fit(X_train, y_train) #training the algorithm
```

```python
#To retrieve the intercept:
print(regressor.intercept_)
#For retrieving the slope:
print(regressor.coef_)
```

```python
y_pred = regressor.predict(X_test)
```

```python
prediction = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
prediction
```

```python
pred_plot = prediction.head(30)
pred_plot.plot(kind='bar',figsize=(16,10))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

### Errors

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
