```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## 2. Fill missing entries (NaNs) in the numerical columns with 0

```python
# Dropping the categorical columns & filling the missing entries in the numerical columns with 0

new_mouse = mouse.drop(columns=['MouseID','Genotype','Treatment','Behavior','class'])
new_mouse = new_mouse.fillna(0)
new_mouse.head(10)
```

## 3. Subtract the mean from each numerical column (standardize data)

```python
new_mouse.describe()
```

```python
standardize = lambda x: (x-x.mean()) / x.std()
standardizedData = new_mouse.pipe(standardize)
standardizedData.head()
```

## 4. Select top 5 numerical features that separate the groups in the "class" column

```python
ix = pd.get_dummies(mouse,columns=['class']).drop(columns=['MouseID','Treatment','Behavior','Genotype'])
ix.head(10)
```

```python
## We are interested in the top 5 features that best discriminate between the classes
## This is a feature selection problem (as opposed to feature transformation accomplished through PCA)
## Since we didn't have time to cover this during the course, I will give you some hints

from sklearn.feature_selection import mutual_info_classif
from sklearn.feature_selection import SelectKBest

# use the mutual_info_classif criterion which measures the mutual information between a numerical feature and a categorical feature (our class column)
# SelectKBest will compute a discrepancy metric (in our case mutual_info_classif) between each numerical column and the class column and return the top k features with the hightest score

numFeatures = 5
targetData = mouse['class'] # the class column from the dataframe
numericalData = standardizedData # the standardized numerical data from the previous paragraph

# Assuring that we don't have NaN values in numericalData - we should get 0
print('There are {} numerical values'.format(numericalData.isnull().sum().sum()))

best_features = SelectKBest(mutual_info_classif, k=numFeatures).fit_transform(numericalData,targetData)

print("Printing the first 5 best features:\n",best_features[:5])
```

```python
pd.options.display.max_columns = None
display(numericalData.iloc[:].head())

#numericalData.iloc[:].head()

# We notice that columns 1, 10, 32, 46, 76 are the best predictors for our 'class'  column 
# Displaying those specific columns in order to assure that we have the right values corresponding to best_features

numericalData.iloc[:,[1,10,32,46,76]]
```

## 5. Show the scatter matrix for the selected features (with hue = class column) 
## 6. Perform PCA on all standarized numerical features

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

no_components = 77
standardizedFeatures= StandardScaler().fit_transform(numericalData)
pca = PCA(n_components = no_components).fit(standardizedFeatures)

pcaFeatures = pd.DataFrame(pca.transform(standardizedFeatures))
pcaFeatures['class']=targetData

# Plotting for only 5 of the features found above.
sns.pairplot(pcaFeatures,hue="class",vars=pcaFeatures.columns[[1,10,32,46,76]])
```

## 7. Using a bar plot, show the variance explained by the first k principal components (with k = 1 to numberOfFeatures)

```python
figure = plt.figure(figsize = (40,20))
x = np.arange(1,no_components+1,1)
height = pca.explained_variance_ratio_.cumsum()
plt.xlabel('Principal component',fontdict = {'fontsize':30}) 
plt.ylabel('Explained variance',fontdict = {'fontsize':30})

plt.bar(x, height)
for i, v in enumerate(height):
    plt.text(x[i] - 0.25, v + 0.01, r'%.2f'%v)
```
